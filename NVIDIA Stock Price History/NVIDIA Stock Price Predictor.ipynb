```python
import pandas as pd
import numpy as np
import datetime

# Helper function to convert string dates to datetime format
def str_to_datetime(date_str):
    return datetime.datetime.strptime(date_str, '%Y-%m-%d')

# Function to convert a DataFrame to a windowed DataFrame
def df_to_windowed_df(dataframe, first_date_str, last_date_str, n=3):
    first_date = str_to_datetime(first_date_str)
    last_date = str_to_datetime(last_date_str)

    target_date = first_date

    dates = []
    X, Y = [], []

    last_time = False
    while True:
        # Subset of the dataframe up to the current target date
        df_subset = dataframe.loc[:target_date].tail(n+1)

        if len(df_subset) != n+1:
            print(f'Error: Window of size {n} is too large for date {target_date}')
            return

        # Extract the features (x) and target (y) from the subset
        values = df_subset['Close'].to_numpy()
        x, y = values[:-1], values[-1]

        # Append to results
        dates.append(target_date)
        X.append(x)
        Y.append(y)

        # Find the date for the next week
        next_week = dataframe.loc[target_date:target_date + datetime.timedelta(days=7)]
        next_datetime_str = str(next_week.head(2).tail(1).index.values[0])
        next_date_str = next_datetime_str.split('T')[0]
        year, month, day = next_date_str.split('-')
        next_date = datetime.datetime(day=int(day), month=int(month), year=int(year))

        # Stop if we have reached the last date
        if last_time:
            break

        # Update the target date
        target_date = next_date

        # Check if the next target date is the last date
        if target_date == last_date:
            last_time = True

    # Create a new DataFrame with the collected data
    ret_df = pd.DataFrame({})
    ret_df['Target Date'] = dates

    # Convert the features (X) to a numpy array
    X = np.array(X)

    # Create columns for each feature
    for i in range(0, n):
        ret_df[f'Target-{n-i}'] = X[:, i]

    # Add the target column
    ret_df['Target'] = Y

    return ret_df

# Assuming `df` is your DataFrame and you want to create a windowed DataFrame
windowed_df = df_to_windowed_df(df, '1999-01-27', '2024-09-20', n=3)
```

```python
def windowed_df_to_date_X_y(windowed_dataframe):
    df_as_np = windowed_dataframe.to_numpy()

    dates = df_as_np[:, 0]

    middle_matrix = df_as_np[:, 1:-1]
    X = middle_matrix.reshape((len(dates), middle_matrix.shape[1], 1))

    Y = df_as_np[:, -1]

    return dates, X.astype(np.float32), Y.astype(np.float32)

dates, X, y = windowed_df_to_date_X_y(windowed_df)

dates.shape, X.shape, y.shape
```
