```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
#Dividing the data into training and testing data

from sklearn.model_selection import train_test_split

X = data.drop(["median_house_value"], axis = 1)
y = data["median_house_value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
#Altering the distribution to produce more Gaussian-like distributions

train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["households"] = np.log(train_data["households"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
#Encoding the proximity to binary values

train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
train_data["bedroom_ratio"] = train_data["total_bedrooms"] / train_data["total_rooms"]
train_data["houshold_rooms"] = train_data["total_rooms"] / train_data["households"]
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, y_train = train_data.drop(["median_house_value"], axis = 1), train_data["median_house_value"]
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s, y_train)
```

```python
test_data = X_test.join(y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)

test_data["bedroom_ratio"] = test_data["total_bedrooms"] / test_data["total_rooms"]
test_data["houshold_rooms"] = test_data["total_rooms"] / test_data["households"]

X_test, y_test = test_data.drop(["median_house_value"], axis = 1), test_data["median_house_value"]
```

```python
test_data
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, y_test)
```
